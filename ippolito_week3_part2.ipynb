```python
# Code taken from https://networkx.org/documentation/stable/auto_examples/algorithms/plot_davis_club.html
import matplotlib.pyplot as plt
import networkx as nx
import networkx.algorithms.bipartite as bipartite

G = nx.davis_southern_women_graph()
women = G.graph["top"]
clubs = G.graph["bottom"]

print("Biadjacency matrix")
print(bipartite.biadjacency_matrix(G, women, clubs))

# project bipartite graph onto women nodes
W = bipartite.projected_graph(G, women)

print()
print("#Friends, Member")
for w in women:
    print(f"{W.degree(w)} {w}")

# project bipartite graph onto women nodes keeping number of co-occurence
# the degree computed is weighted and counts the total number of shared contacts
W = bipartite.weighted_projected_graph(G, women)
print()
print("#Friend meetings, Member")
for w in women:
    print(f"{W.degree(w, weight='weight')} {w}")

pos = nx.spring_layout(G, seed=648)  # Seed layout for reproducible node positions
nx.draw(G, pos, with_labels=True)
plt.show()
```

```python
# Import extra libraries
import plotly.express as px
import plotly.io as pio
pio.renderers.default = 'jupyterlab'    # set ploty renderer so graphs show up in HTML output

# Draw projected graph of women
pos = nx.spring_layout(W, seed=648)  # Seed layout for reproducible node positions
nx.draw(W, pos, with_labels=True, alpha=0.5)
plt.show()
```

```python
# Degree centrality - women

# Extract women and degrees sorted in descending fashion by degree
srt = sorted(W.degree(weight='weight'), key=lambda tup: tup[1], reverse=True)
w = [w[0] for w in srt]
d = [w[1] for w in srt]

# List the women by weighted degree (the "celebrities") and display vertical bar chart
print(srt)
p2 = px.bar(x=w, y=d, title='Weighted degree')
p2.show()

# Show histogram of weighted degrees
p1 = px.histogram(d, nbins=8, title='Weighted degree')
p1.update_layout(showlegend=False) 
p1.show()
```

We see that Theresa Anderson had the highest weighted degree, meaning she was associated to the most clubs frequented by other women in the sample group. The histogram shows a normal distribution of weighted degrees.


```python
# Closeness centrality - women

# Extract women and closeness centrality sorted in descending fashion by closeness centrality
srt = sorted(nx.closeness_centrality(W).items(), key=lambda item: item[1], reverse=True)
w = [w[0] for w in srt]
cc = [w[1] for w in srt]
print(srt)

# List the women by weighted closeness centrality (the "gossip mongers") and display vertical bar chart
p2 = px.bar(x=w, y=cc, title='Weighted closeness centrality')
p2.show()

# Show histogram of weighted degrees
p1 = px.histogram(cc, nbins=8, title='Weighted closeness centrality')
p1.update_layout(showlegend=False) 
p1.show()
```

Based on the weighted closeness centrality statistics, it is apparent that a number of women (six of them) share the top rank in terms of who the gossip mongers are.


```python
# Betweenness centrality - women

# Extract women and betweenness centrality sorted in descending fashion by betweenness centrality
srt = sorted(nx.betweenness_centrality(W, weight='weight').items(), key=lambda item: item[1], reverse=True)
w = [w[0] for w in srt]
bc = [w[1] for w in srt]
print(srt)

# List the women by weighted closeness centrality (the "gossip mongers") and display vertical bar chart
p2 = px.bar(x=w, y=bc, title='Weighted betweenness centrality')
p2.show()

# Show histogram of weighted degrees
p1 = px.histogram(bc, nbins=8, title='Weighted betweenness centrality')
p1.update_layout(showlegend=False) 
p1.show()
```

Olivia Carleton and Flora Price are the standouts for betweenness centrality, indicating that they may form bridges (or bottlenecks) among different groups.


```python
# Eigenvector centrality - women

# Extract women and eigenvector centrality sorted in descending fashion by eigenvector centrality
srt = sorted(nx.eigenvector_centrality(W, weight='weight').items(), key=lambda item: item[1], reverse=True)
w = [w[0] for w in srt]
ec = [w[1] for w in srt]
print(srt)

# List the women by weighted closeness centrality (the "gossip mongers") and display vertical bar chart
p2 = px.bar(x=w, y=ec, title='Weighted eigenvector centrality')
p2.show()

# Show histogram of weighted degrees
p1 = px.histogram(ec, nbins=8, title='Weighted eigenvector centrality')
p1.update_layout(showlegend=False) 
p1.show()
```

Theresa Anderson seems to be the "gray cardinal" of the bunch, with the highest eigenvector centrality of 0.36.


```python
# Until this point we've looked at the network from the perspective of how the women were connected to each other.
# Now look at how the clubs are linked.

# Project bipartite graph onto club nodes
C = bipartite.weighted_projected_graph(G, clubs)

# Print club memberships
print("#Members, Club")
for c in clubs:
    print(f"{C.degree(c, weight='weight')} {c}")

# Draw projected graph of clubs
pos = nx.spring_layout(C, seed=648)  # Seed layout for reproducible node positions
nx.draw(C, pos, with_labels=True, alpha=0.5)
plt.show()
```

```python
# Degree centrality - clubs

# Extract clubs and degrees sorted in descending fashion by degree
srt = sorted(C.degree(weight='weight'), key=lambda tup: tup[1], reverse=True)
c = [w[0] for w in srt]
d = [w[1] for w in srt]

# List the clubs by weighted degree (the "celebrities") and display vertical bar chart
print(srt)
p2 = px.bar(x=c, y=d, title='Weighted degree')
p2.show()

# Show histogram of weighted degrees
p1 = px.histogram(d, nbins=8, title='Weighted degree')
p1.update_layout(showlegend=False) 
p1.show()
```

With a weighted membership of 59, E8 is the most popular club.


```python
# Analyze subcomponents

# Get number of subgraphs - women
print("Number of women's subgraphs:", len([c for c in nx.connected_components(W)]))

# Get number of subgraphs - clubs
print("Number of club subgraphs:", len([c for c in nx.connected_components(C)]))
```

No disconnected networks exist in either the case of women or of clubs.


```python
# Define function to trim edges based on weight
# Maksim Tsvetovat & Alexander Kouznetsov, Social Network Analysis for Startups, 2011
def trim_edges(g, weight=1):
    g2 = nx.Graph()
    for f, to, edata in g.edges(data=True):
        if edata['weight'] > weight:
            g2.add_edge(f, to, weight=edata['weight'])
    return g2
```

```python
# Find subgraphs
#print(sorted(W.edges(data=True), key=lambda wt: wt[2]['weight'], reverse=True))
for i in range(1, 7):
    print(i)
    WW = trim_edges(W, i)
    print("weight=" + str(i) + ":", len([c for c in nx.connected_components(WW)]), "subgraphs")
```

The fact that there are at most two subgraphs indicates that the women are all well connected as opposed to disjointed.
